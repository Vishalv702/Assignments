```python
# 1. Create a set of five city names and print the set.
city = {"pune","mumbai","satara","buldhana","chikhli"}
print(city)
```

```python
# 2. Add multiple elements into a set using update().
city = {"pune","mumbai","satara","buldhana","chikhli"}
city.update(["kolhapur","raigad"])
print(city)
```

```python
# 3. Remove an element from a set using remove() and discard().
city = {"pune","mumbai","satara","buldhana","chikhli"}
city.remove("pune")
print(city)

city.discard("satara")
print(city)
```

```python
# 4. Find the union of two sets.
city = {"pune","mumbai","satara","buldhana","chikhli"}
city_2 = {"kolhapur","raigad"}
print("union of sets: ", city | city_2)
```

```python
# 5. Find the intersection of two sets.
city = {"pune","mumbai","satara","buldhana","chikhli"}
city_2 = {"kolhapur","raigad","pune","chikhli"}
intersection = city & city_2
print("Intersection of sets: ", intersection)
```

```python
# 6. Find the difference between two sets.
A = {22,4,67,52,19}
B = {52,38,4}
diff = A - B 
print(diff)
```

```python
# 7. Check whether a particular element exists in a set.
num = {22,4,67,52,19}
print("Is 22 in num: ",22 in num)
```

```python
# 8. Convert a list with duplicate values into a set.
num = [22,111,22,33,44,33,77]
unique = set(num)
print(unique)
```

```python
# 9. Create a frozen set and explain why elements cannot be modified.
#  frozen set is immutable , means it cant be change once declared
frozen = frozenset([12,13,14,15])
print("Frozen set: ",frozen)
```

```python
# 10. Perform symmetric difference operation on two sets.
set1 = {1,2,3,4}
set2 = {3,4,5,6}

result = set1.symmetric_difference(set2)

print(result)
```
