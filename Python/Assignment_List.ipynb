```python
# 14. Reverse a given list without using the reverse() function.
numbers = [1,2,35,46,51,68,73,89,94,0] 
length=len(numbers)
for i in range(length-1,-1,-1):
   print(numbers[i],end=" ")
```

```python
# 15. Remove duplicate elements from a list.
numbers = [1,23,35,46,35,68,73,23,94,0]
unique=[]

for i in numbers:
    if i not in unique:
        unique.append(i)
print(unique)
```

```python
# 7. Find the second largest number in a list.
numbers = [1,23,35,46,35,68,73,23,94,0]
max_num=0
s_max=0
for i in numbers:
    if i > max_num:
        s_max=max_num
        max_num=i
    if i >s_max and i != max_num:
        s_max=i

print(s_max)        
        
```

```python
# 8. Count how many times a specific element appears in a list.
numbers = [1,23,35,46,35,68,73,23,94,1]
num=23
count =0;
for i in numbers:
    if i == num:
        count +=1
print(count)        
        
```

```python
# 10. Convert a list of strings into uppercase using a loop.
students = ["Vishal","Vedant","Arnav","Om"]
length = len(students)
for i in range(length):
    students[i] = students[i].upper()
print(students)    
```
