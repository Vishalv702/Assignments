```python

# 12. Create a dictionary of employees with their departments.
employees = {
    "Vishal": "IT",
    "Vedant": "HR",
    "Arnav": "Finance"
}
```

```python

# 13. Access the department of a specific employee.
employees["Vedant"]
```

```python

# 14. Add a new employee and department.
employees.update({"ashwini" : "sales"})
print(employees)
```

```python

# 15. Change the department of an existing employee.
employees.update({"Vedant":"Marketing"})
print(employees)
```

```python

# 16. Remove an employee from the dictionary.
del employees["ashwini"]

print(employees)
```

```python
# 17. Sort the employee dictionary by employee names in ascending order.
sort_by_name = dict( sorted( employees.items() ) ) 
print("Sort by emplyee name: ", sort_by_name ) 
```

```python
# 18. Sort the employee dictionary by department names in ascending order.
sort_by_dept = dict( sorted( employees.items() , key = lambda x : x[1] ) ) 
print("Sort by emplyee department: ", sort_by_dept ) 
```

```python
# 19. Sort the employee dictionary by names in descending order.
sort_by_name = dict( sorted( employees.items() , reverse =True ) ) 
print("Sort by emplyee name: ", sort_by_name ) 
```

```python
# 20. Sort the employee dictionary by department in descending order.
sort_by_dept = dict( sorted( employees.items(), key = lambda x:x[1] , reverse = True ) ) 
print("Sort by emplyee department: ", sort_by_dept ) 
```

```python
# 21. Find all keys that have a specific value in a dictionary.

students = { "vishal":83 , "ram":71 , "shaym":94 , "priya":40, "riya":83 }
keys_list =[]
for key , value in students.items():
    if value == 83:
        keys_list.append(key)
        
print(keys_list)        
```

```python
# 22. Count how many students scored above 80 marks.
count = 0 
for value in students.values():
    if value >=80:
        count += 1

print(count)
```

```python
# 23. Create a nested dictionary for 3 students with age, grade, and department.

students = {
    "Vishal": {
        "age": 22,
        "grade": "A",
        "department": "CS"
    },

    "Vedant": {
        "age": 21,
        "grade": "B",
        "department": "IT"
    } ,
         "Ram": {
        "age": 23,
        "grade": "A",
        "department": "CE"
    }
}

print(students)
```

```python
# 24. Access the grade of a student from the nested dictionary.

students["Vishal"]["age"]
```

```python

# 25. Update the department of a student in the nested dictionary.

students["Vedant"]["department"] = "EE"

students["Vedant"]["department"]
```

```python
# 26. Loop through the nested dictionary and print student names with their grades.

for name , details in students.items():
    
    for key , value in details.items():
        if key =="age":
          print("Name: ",name, "age: ", value)
```
